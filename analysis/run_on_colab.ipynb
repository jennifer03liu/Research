```python
# Step 4: Zip the results folder for download
import os
import shutil
from google.colab import files

# Find the latest results directory
dirs = [d for d in os.listdir('.') if os.path.isdir(d) and '_Results' in d]
dirs.sort(key=lambda x: os.path.getmtime(x), reverse=True)

if dirs:
    latest_dir = dirs[0]
    print(f"Zipping {latest_dir}...")
    shutil.make_archive(latest_dir, 'zip', latest_dir)
    print("Downloading zip file...")
    files.download(f"{latest_dir}.zip")
else:
    print("No results directory found.")
```
